```python
# importing necessary modules and libraries

from pandas_profiling import ProfileReport 
import random
import json
import csv
import math
import pandas as pd
import numpy as np
```

```python
# declaring a dictionary which will contain all the excelbooks that we are working with

dataframe=[]
df11=pd.read_excel('../my_csv/fixtures/fixtures 11-12.xlsx')
dataframe.append(df11)
df12=pd.read_excel('../my_csv/fixtures/fixtures 12-13.xlsx')
dataframe.append(df12)
df13=pd.read_excel('../my_csv/fixtures/fixtures 13-14.xlsx')
dataframe.append(df13)
df14=pd.read_excel('../my_csv/fixtures/fixtures 14-15.xlsx')
dataframe.append(df14)
df15=pd.read_excel('../my_csv/fixtures/fixtures 15-16.xlsx')
dataframe.append(df15)
df16=pd.read_excel('../my_csv/fixtures/fixtures 16-17.xlsx')
dataframe.append(df16)
df17=pd.read_excel('../my_csv/fixtures/fixtures 17-18.xlsx')
dataframe.append(df17)
df18=pd.read_excel('../my_csv/fixtures/fixtures 18-19.xlsx')
dataframe.append(df18)
df19=pd.read_excel('../my_csv/fixtures/fixtures 19-20.xlsx')
dataframe.append(df19)
df20=pd.read_excel('../my_csv/fixtures/fixtures 20-21.xlsx')
dataframe.append(df20)
df21=pd.read_excel('../my_csv/fixtures/fixtures 21-22.xlsx')
dataframe.append(df21)
```

```python
# drop any NaN values that may be present in the dataframes

for count in dataframe:
    count.dropna(inplace=True)
```

```python
# creating common_dict that will contain the different teams in all the excelbooks and the number of matches they have played in the past 10 years
# sorting this dictionary in descending order according to matches played and then adding the top 20 teams to common_list

common_dict={}
for count in dataframe:
    for skip in count['Home']:
        if skip not in common_dict:
            common_dict[skip]=1
        else:
            common_dict[skip]+=1
common_list=[]
common_dict=sorted(common_dict.items(),key=lambda x:x[1],reverse=True)
for count in range(20):
    common_list.append(common_dict[count][0])
```

```python
# creating dataframe with all teams from common_list, with columns for W,D,L at home and away and filling all columns with 0

new=pd.DataFrame(columns=['Team','HW','HD','HL','AW','AD','AL'])
new['Team']=common_list
new.fillna(0,inplace=True)
new
```

```python
# filling columns in 'new' dataframe with relevant values 

# traverse through teams in common_list
for count in common_list:
    # traverse through each year's schedule and results in 'dataframe'
    for skip in dataframe:
        # going row by row in above schedule
        for jump in skip.index:
            # wins at home
            if skip['Home'][jump]==count and skip['Score'][jump][0]>skip['Score'][jump][2]:
                new['HW'][common_list.index(count)]+=1
            # wins away
            elif skip['Away'][jump]==count and skip['Score'][jump][2]>skip['Score'][jump][0]:
                new['AW'][common_list.index(count)]+=1
            # draws at home
            elif skip['Home'][jump]==count and skip['Score'][jump][0]==skip['Score'][jump][2]:
                new['HD'][common_list.index(count)]+=1
            # draws away
            elif skip['Away'][jump]==count and skip['Score'][jump][2]==skip['Score'][jump][0]:
                new['AD'][common_list.index(count)]+=1
            # losses at home
            elif skip['Home'][jump]==count and skip['Score'][jump][0]<skip['Score'][jump][2]:
                new['HL'][common_list.index(count)]+=1
            # losses away
            elif skip['Away'][jump]==count and skip['Score'][jump][2]<skip['Score'][jump][0]:
                new['AL'][common_list.index(count)]+=1
```

```python
# creating columns for total W,D,L along with probability to W,D,L at home and away

new['W']=new['HW']+new['AW']
new['D']=new['HD']+new['AD']
new['L']=new['HL']+new['AL']
HG=new['HW']+new['HD']+new['HL']
AG=new['AW']+new['AD']+new['AL']
new['hgwin']=round(new['HW']/HG,3)
new['hgdraw']=round(new['HD']/HG,3)
new['hgloss']=round(new['HL']/HG,3)
new['agwin']=round(new['AW']/AG,3)
new['agdraw']=round(new['AD']/AG,3)
new['agloss']=round(new['AL']/AG,3)
new.sort_values('W',ascending=False,inplace=True)
new.reset_index(drop=True,inplace=True)
new
```
